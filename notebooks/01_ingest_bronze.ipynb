#### General code for reading a file from a path and then doing a full load

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

input_path = ""

#read the csv file from a path
df = spark.read.format("csv").option("header", True).option("inferschema", True).load(input_path)

#writing the data as it is
df.write.format("delta").mode("overwrite").saveAsTable("finance_catalog.bronze.creditcard_bronze")
```
